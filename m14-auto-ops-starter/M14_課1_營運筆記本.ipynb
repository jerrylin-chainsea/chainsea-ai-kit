## 1. 正常營業日:今天賣了什麼(不用金鑰)
讀 M12 的範例資料,算每個品項總量、標出怪的 👇

```python
import csv, collections
rows=list(csv.DictReader(open('../m12-crew-starter/data/sales_sample.csv',encoding='utf-8')))
tot=collections.Counter()
for r in rows:
    q=int(r['qty']); tot[r['item']]+=q
    if q<0 or q>200: print(f'⚠ 怪資料:{r["date"]} {r["item"]} 數量={q}')
print('\n各品項總量:')
for it,q in tot.items(): print(f'  {it:8s} {q}')
```

## 2. 一年史:有 AI vs 沒 AI(概念模擬)
AI 第 6 個月介入,毛利走向就此岔開 👇

```python
base=60  # 起始毛利%
no_ai=[max(30, base-i*2) for i in range(12)]          # 沒 AI:一路往下
with_ai=[max(30, base-i*2) if i<6 else 36+(i-5)*4 for i in range(12)]  # 有 AI:第6月回升
print('月  沒AI        有AI')
for i in range(12):
    bar_n='█'*(no_ai[i]//4); bar_a='█'*(with_ai[i]//4)
    star=' ← AI 介入' if i==6 else ''
    print(f'{i+1:2d}  {no_ai[i]:3d}{bar_n:<16s} {with_ai[i]:3d}{bar_a}{star}')
```
