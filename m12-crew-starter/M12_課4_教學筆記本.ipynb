# M12 · 課4 ｜ 可信任規則 + 三情境測試(動手)

別只測『會成功』的。測三種:正常 / 缺資料 / 錯誤輸入。

## 1. 三情境測試(不用金鑰,純規則跑)
正常放行;缺資料、錯誤要擋下要人確認 👇

```python
def check(rows):
    if not rows: return '擋下:資料不齊,需要人確認'
    for r in rows:
        if r['qty'] < 0: return f'擋下:偵測到負值 {r}，要人確認'
        if r['price'] < r['cost']: return f'擋下:售價低於成本 {r}，要人確認'
    return '放行:資料正常,可交給 AI 分析'
scenarios = {
  '正常':   [{'item':'原味','qty':120,'price':5,'cost':3}],
  '缺資料': [],
  '錯誤':   [{'item':'原味','qty':-5,'price':5,'cost':3}],
}
for name, rows in scenarios.items(): print(f'{name:5s} → {check(rows)}')
assert '放行' in check(scenarios['正常']) and '擋下' in check(scenarios['缺資料'])
print('\n✓ 正常放行、缺漏與錯誤擋下 = 把關生效')
```

## 2. 換你埋雷
加一筆『售價低於成本』,看它抓不抓 👇

```python
trap=[{'item':'原味','qty':50,'price':1,'cost':3}]  # 售價 1 < 成本 3
print(check(trap))
```
