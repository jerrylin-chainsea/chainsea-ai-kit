## 2. 換資料(程式示範:把某筆改成誇張值)

```python
rows2=[r[:] for r in rows]
rows2[3][3]='9999'  # 把某筆數量改成 9999
print('改後第 3 筆:', rows2[3], '← 等下跑 AI 它會抓這筆異常')
```
